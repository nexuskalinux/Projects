```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split # version 0.18.1
# split the data with 50% in each set
# la fonction split permet de supprimer des valeurs d'un tableau, en l'occurence 50% pour chacun de nos tableau

data_test = train_test_split(data, target
                                 , random_state=0 # test de façon aléatoire "random"
                                 , train_size=0.5)
data_train, data_test, target_train, target_test = data_test
model = LogisticRegression()
model = model.fit(data_train, target_train)
result = model.predict(data_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(result, target_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
sns.heatmap(confusion_matrix(target_test, result), square=True, annot=True, cbar=False
            , xticklabels=list(iris.target_names)
            , yticklabels=list(iris.target_names))
plt.xlabel('valeurs prédites')
plt.ylabel('valeurs réelles');
```
