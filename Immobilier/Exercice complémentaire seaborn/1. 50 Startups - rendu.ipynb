# 50 Startups

Echauffons nous sur la visualisation avec un dataset simple. Celui-ci décrit le profit de 50 startups en fonction de leurs dépenses dans différents départements. 

**1. Importez le dataset & visualisez le**

```python
import pandas as pd
import seaborn as sns
sns.set(style="ticks", rc={'figure.figsize':(5.5,7)})
```

```python
startups = pd.read_csv("./50_Startups.csv")
startups.head()
```

**2. Avez-vous des valeurs NaN dans le dataset ?**

```python
"""
startups.isnull().sum()
startups.isnull().values.any()
startups.isna().sum()
startups.isna().sum().sum()
"""
startups.isnull().any()
```

**3. Faites un nuage de points qui va vous permettre d'exprimer le profit en fonction des dépenses en R&D**

```python
sns.relplot(x="R&D Spend", y="Profit", data=startups)
```

**4. Faites de même avec les autres variables quantitatives**

```python
sns.relplot(x="Administration", y="Profit", data=startups)
```

```python
sns.relplot(x="Marketing Spend", y="Profit", data=startups)
```

**5. Regardez maintenant la part de profit en fonction des pays. Vous pourrez l'exprimer par un barplot puis un boxplot. Quel graph est plus pertinent selon vous ?**

```python
sns.set(style="ticks", rc={'figure.figsize':(5,6)})
sns.barplot(y="Profit", x="State", data=startups)
```

```python
sns.boxplot(y="Profit", x="State", data=startups)
```

**7. Faites une heatmap qui va vous permettre de regarder la corrélation entre les variables. Que pouvez-vous conclure sur les variables les plus influentes sur le profit ?**

```python
sns.set(style="ticks", rc={'figure.figsize':(6,5)})
sns.heatmap(startups.corr(), annot=True)
```
